# Data Exploration

```python
# Importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Reading in the data

df = pd.read_csv(r'/Users/shrutidua/Downloads/world_population.csv')
df
```

```python
# Changing the scientific notation

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Looking at the information

df.info()
```

```python
# Overview of the columns

df.describe()
```

```python
# Finding missing values

df.isnull().sum()
```

```python
# Number of unique values

df.nunique()
```

```python
# Looking at the top 10 countries with the lasrgest world population percentage

df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
# Looking at the Correlation Matrix between the columns. We are using the Pearson method.

df.corr(method ='pearson', numeric_only = True)
```

```python
# Visualising the correlation matrix using a heatmap

sns.heatmap(df.corr(method ='pearson', numeric_only = True), annot = True)

plt.show()
```

```python
# Grouping the data by continent

df.groupby('Continent').mean(numeric_only = True).sort_values(by="2022 Population",ascending=False)
```

```python
# Continent populations over the years 1970 - 2022

continent_population = df.groupby('Continent')[['1970 Population', '1980 Population', '1990 Population', 
                                                '2000 Population', '2010 Population', '2015 Population', 
                                                '2020 Population', '2022 Population']].mean(numeric_only = True).sort_values(by="2022 Population",ascending=False)

continent_population = continent_population.transpose()

continent_population
```

```python
#Visualising the continent populations over the years 1970 - 2022

continent_population.plot()

plt.title('Continent Population Over the Years 1970 - 2022')
plt.xlabel('Year')
plt.ylabel('Population')
```

```python
# Comparing the average population density grouped by the continents

continent_density = df.groupby('Continent')['Density (per km²)'].mean(numeric_only = True).sort_values(ascending=False)

continent_density
```

```python
# Visualising population density comparison of continents

plt.bar(continent_density.index, continent_density.values)

plt.title('Average Population Density by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Population Density (per km²)')

plt.xticks(rotation=45)
plt.show()
```
