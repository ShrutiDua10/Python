# Data Cleaning using Pandas

```python
# In this project, I am cleaning a small random dataset using Pandas to demonstrate the data cleaning process
```

```python
import pandas as pd
```

```python
# Reading in the data

df = pd.read_excel(r"/Users/shrutidua/Documents/Pandas/Customer Call List.xlsx")
df
```

```python
# Dropping any duplicates

df = df.drop_duplicates()
df
```

```python
# Dropping any columns that are not needed

df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Last_Name

# Stripping the last name of any unwanted characters

df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# Phone_Number

# Getting rid of non-numeric and non-alphabetic characters in the phone numbers

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
# Converting the phone numbers into strings

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
# Standardizing the phone number format with hyphens

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
df
```

```python
# Getting rid of unavailable phone numbers

df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", '')
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", '')
df
```

```python
# Address

# Splitting the address to make it more clear and standardized

df[["Street_Address", "State", "Zipcode"]] = df["Address"].str.split(',',n = 2, expand = True)
df
```

```python
# Paying Customer and Do_Not_Contact

# Standardizing the No/Yes Data

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N') 
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y') 
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N') 
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df
```

```python
# Removing null values 

df = df.fillna('')
df
```

```python
# Dropping rows with a "Yes" (Y) response to Do_Not_Contact

for i in df.index:
    if df.loc[i, "Do_Not_Contact"] == "Y":
        df.drop(i, inplace = True)
        
df
```

```python
# Dropping rows with no phone number given

for i in df.index:
    if df.loc[i, "Phone_Number"] == '':
        df.drop(i, inplace = True)
        
df
```

```python
# Resetting the index of the cleaned final data

df = df.reset_index(drop = True)
df
```
